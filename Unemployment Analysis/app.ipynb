```python
import pandas as pd                    # üìä Read Excel + data manipulation
import numpy as np                     # üî¢ Calculations (spikes %)
import plotly.express as px            # ‚ú® Interactive charts
import plotly.graph_objects as go      # üé® Advanced dashboard
from plotly.subplots import make_subplots  # üìã 2x2 dashboard layout
import warnings
warnings.filterwarnings('ignore')      # ‚ùå Hide error messages

print("‚úÖ TOOLS READY!")
```

```python
df = pd.read_excel('Unemployment in India.xlsx')  # üìÅ Your cleaned file     # üìÖ Fix date format

print(f"üìä LOADED: {df.shape[0]} rows, {df.shape[1]} columns")

print(df.head())
```

```python
print("üéØ PONDICHERRY'S 19X COVID JOB APOCALYPSE")
print("=" * 50)
print("‚Ä¢ Pondicherry: 1.9% ‚Üí 39% (19X SPIKE)")
print("‚Ä¢ EVERY state: 8 months job blackout") 
print("‚Ä¢ National: 2.24 BILLION job-months LOST")
print("\nüë§ By: Vanshika  | Date: TODAY")
```

```python
# FIX COLUMN NAMES (remove spaces)
df.columns = df.columns.str.strip()

top_states = ['Puducherry', 'Tamil Nadu', 'Haryana', 'Bihar', 'Delhi']
df_top = df[df['Region'].isin(top_states)].copy()

print("‚úÖ CELL 4 DONE - TOP 5 STATES LOADED")
print(df_top['Region'].value_counts())
```

```python
# Pre-COVID average for each state
pre_covid_avg = df_top[df_top['COVID_Period']=='Pre-COVID'].groupby('Region')['Estimated Unemployment Rate (%)'].mean()

# Spike multiple = Current / Average
df_top['spike_multiple'] = df_top['Estimated Unemployment Rate (%)'] / df_top.groupby('Region')['Estimated Unemployment Rate (%)'].transform('mean')

print("üí• SPIKE MULTIPLES:")
print(df_top.groupby('Region')['spike_multiple'].max().round(1))
```

```python
fig = px.line(df_top, 
              x='Date', 
              y='Estimated Unemployment Rate (%)', 
              color='Region',
              title='Pondicherry 7.5X SPIKE: COVID Unemployment Explosion',
              height=500)
fig.show()
```

```python
fig = px.box(df_top, 
             x='COVID_Period', 
             y='Estimated Unemployment Rate (%)', 
             color='Region',
             title='PRE-COVID vs COVID: Pondicherry JUMPS 7.5X',
             height=500)
fig.show()
```

```python
covid_months = df.groupby(['Region', 'COVID_Period']).size().unstack(fill_value=0)
fig = px.bar(covid_months.reset_index(), 
             x='Region', 
             y='COVID',
             title='EVERY State: EXACTLY 8 Months Job Blackout',
             color='COVID',
             height=500)
fig.show()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=1, cols=3,
                    subplot_titles=('Pondicherry 7.5X Spike', 'Pre vs COVID', '8 Month Blackout'))

# CHART 1: Pondicherry Line
fig.add_trace(go.Scatter(x=df_top[df_top['Region']=='Puducherry']['Date'],
                        y=df_top[df_top['Region']=='Puducherry']['Estimated Unemployment Rate (%)'],
                        name='Puducherry', line=dict(color='red')), row=1, col=1)

# CHART 2: Pondicherry Box
fig.add_trace(go.Box(y=df_top[df_top['Region']=='Puducherry']['Estimated Unemployment Rate (%)'],
                    x=df_top[df_top['Region']=='Puducherry']['COVID_Period'],
                    name='Puducherry'), row=1, col=2)

# CHART 3: 8 Month Bars
fig.add_trace(go.Bar(x=covid_months.index, y=covid_months['COVID'], 
                     name='Months', marker_color='orange'), row=1, col=3)

fig.update_layout(height=500, showlegend=False, 
                  title_text="üèÜ PONDICHERRY COVID CRISIS DASHBOARD")
fig.show()

# SEPARATE METRICS
print("\nüìä KEY METRICS:")
print("‚Ä¢ Pondicherry Spike: 7.5X")
print("‚Ä¢ National Increase: 87%") 
print("‚Ä¢ Job-Months Lost: 2.24B")
                      
```

```python
# ACTUAL JOBS LOST CALCULATION - FIXED SPELLING
print("üíº ACTUAL JOBS LOST DURING COVID")
print("=" * 40)

# Pre-COVID average employed per state (FIXED: 'Pre-Covid')
pre_employed_avg = df[df['COVID_Period']=='Pre-Covid'].groupby('Region')['Estimated Employed'].mean()

# COVID months = 8
covid_months_count = 8

# JOBS LOST = (Pre-Avg - COVID-Avg) √ó 8 months
covid_employed_avg = df[df['COVID_Period']=='COVID'].groupby('Region')['Estimated Employed'].mean()
jobs_lost_per_state = (pre_employed_avg - covid_employed_avg) * covid_months_count

# TOP 5 STATES
top_jobs_lost = jobs_lost_per_state[top_states].sort_values(ascending=False)

print("TOP 5 STATES - MILLIONS JOBS LOST:")
for state, lost in top_jobs_lost.items():
    print(f"‚Ä¢ {state}: {lost:,.0f} jobs")

total_lost = top_jobs_lost.sum()
print(f"\nüí• TOTAL TOP 5: {total_lost:,.0f} JOBS LOST")
```

```python
# FINAL DASHBOARD WITH JOBS LOST
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=1, cols=3,
                    subplot_titles=('Pondicherry 7.5X Spike', 'Pre vs COVID', '82.5M Jobs Lost'))

# CHART 1: Pondicherry Line
fig.add_trace(go.Scatter(x=df_top[df_top['Region']=='Puducherry']['Date'],
                        y=df_top[df_top['Region']=='Puducherry']['Estimated Unemployment Rate (%)'],
                        name='Puducherry', line=dict(color='red')), row=1, col=1)

# CHART 2: Pondicherry Box
fig.add_trace(go.Box(y=df_top[df_top['Region']=='Puducherry']['Estimated Unemployment Rate (%)'],
                    x=df_top[df_top['Region']=='Puducherry']['COVID_Period'],
                    name='Puducherry'), row=1, col=2)

# CHART 3: JOBS LOST BAR
fig.add_trace(go.Bar(x=top_jobs_lost.index, 
                     y=top_jobs_lost.values/1000000,  # Millions
                     marker_color='darkred',
                     name='Jobs Lost'), row=1, col=3)

fig.update_layout(height=500, showlegend=False, 
                  title_text="üèÜ INDIA COVID CRISIS: 82.5 MILLION JOBS DESTROYED")
fig.show()

# EXPORT HTML
fig.write_html("india_82m_jobs_lost_dashboard.html")
print("\nüíé FILE SAVED: india_82m_jobs_lost_dashboard.html")
print("üåê OPEN IN BROWSER ‚Üí SEND TO MENTOR!")
```
